Scherer's compliance shorthand recalls Voight notations, so:

|Voight Notation | Direction |
|---------|--------|
|1|xx|
|2|yy|
|3|zz|
|4|yz|
|5|xz|
|6|xy|

So, the notes cover well what happens for the upper left hand side of the matrix? But why are $c_{16}$ and $c_{61}$ non-zero? Well for $\sigma_1$, there is nothing in the $yz$ direction that will trigger an $xx$ response, nor is there a $xy$ factor that will trigger a $zz$ response, so  $c_{16}$ and $c_{26} are non zero, but the others are not.




## Part 2

### Cubic Symmetry:
Since all directions of the \[100\] variety are equivalent, we can say

$$c_{11} = c_{22} = c_{33}$$

and

$$c_{23} = c_{13} = c_{12}$$

so the shear moduli must also be similar:

$$c_{44} = c_{55} = c_{66}$$

So all we need are $c_{11}$, $c_{12}$, and $c_{44}$, and from this we can determine the poisson's ratio.

### Tetragonal Symmetry:
Here we have all 90˚ angles but x=y≠z, so were left with 

$$c_{11} = c_{22} \neq c_{33}$$

and

$$c_{12} ≠ c_{13} = c_{23}$$

and finally 

$$c_{44} = c_{55} \neq c_{66}$$

So all we need are $c_{11}$, $c_{12}$, $c_{13}$, $c_{33},$$c_{44}$,$c_{66}$ and from this we can determine the poisson's ratio.

### Orthorhombic Symmetry:

By the above arguments:
$c_{11}, c_{22}, c_{33}, c_{44}, c_{55}, c_{66}, c_{12}, c_{13}, c_{23}$

### Hexagonal Symmetry:
a≠c, and we have repeating angles at 60 and 90˚, so
$c_{11}$,$c_{12}$,$c_{13}$,$c_{33}$,$c_{44}$

### Rhomohedral Symmetry:

Here a=b=c, but we have varied angle. So we need indepedent vectors for $c_{11}$, $c_{12}$ and $c_{13}$
