## The Ellingham Diagram

(adapted and cribbed from http://web.mit.edu/2.813/www/readings/Ellingham_diagrams.pdf)

Generally when we find metals on earth, they are bound to an oxide. As we discussed over the last few weeks, we generally like metals more than oxides for complex structures because of their mechanical failure-failsafes.

So how do we get rid of the oxide?

Well, generally we can consider

$$M_y+O_x \rightarrow M_yO_x$$

Which proceeds as a reaction as function of temperature where

$$\Delta G = \Delta H - T\Delta S$$

Luckily, just like temperature is high and low, and mechanical forces can be made to a scale, so can reactions.

Meet the Ellingham Diagram



From this chart, we see a lot of different ways in which we can reverse the oxidation of our metal in question


- If we have a metal that is more reducing, we can sacrifice this
- If we used CO, we can use it to futher reduce the metal and make CO_2
- We can pull a vaccuum and push the reaction equilibrium in favor of making the metal at a certain temperature.




What's interesting about reaction _chemical_ reaction kinetics is that they are generally very slow at room temperature (electrochemical reaction kinetics are not), so this means that we can run operations to push the system towards a state where the $M_yO_x$ wants to reduce at high tmeperature, and when we take the metal to room temperature, the drive to oxidize is still there, but for engineering lifetimes it often does not (until electrochemistry (e.g. rust) rears its head).

So we can we do with this?

Let's look at very different bonding structures.

## Polymers

(cribbed heavily from http://www.materials.unsw.edu.au/tutorials/online-tutorials/polymers)

Let's start with our potential well, as we know and love it.

We made two implicit, but very important, assumptions at the beginning of this course that colored the way we saw the world. 

- Within a given structure all bonds have similar weight
- Once one bond forms, all bonds want to form

What happens if these statements are not always true?



Let's start with methane

```
  H
  |
H-C-H
  |
  H
```

It seems like there's not much we can do here, but we if we take two:

```
  H        H
  |        |
H-C-H    H-C-H
  |        |
  H        H
```

Then we can imagine an reaction where

$$2CH_4 \rightarrow  H_2 + C_2H_6$$


Or we can run a reaction where we start with ethelyne

```
H H
| |
C=C
| |
H H
```

And operate on it in such a way to break that double bond:

```
 H H H H H
 | | | | |
-C-C-C-C-C
 | | | | |
 H H H H H 
```

And then it can get even more interesting where:

```
  H   Cl  H   Cl  C   Cl
  |   |   |   |   |   |
- C - C - C - C - C - C -
  |   |   |   |   |   |
  H   H   H   H   H   H

```

Or even


```
                      H
                      |
                  H - C - H
                      |
  H   H   H   H   H   |
  |   |   |   |   |   |
- C - C - C - C - C - C - 
  |   |   |   |   |   |
  H   H   H   H   H   H


```

Or Even

```
                      H
                      |
                  H - C - H
                      |
  H   Cl  H   Cl  H   |
  |   |   |   |   |   |
- C - C - C - C - C - C - 
  |   |   |   |   |   |
  H   H   H   H   H   H


```

And these chains can be designed to almost arbitrary lengths, or _molecular weights_.
