$$F = \frac{E_0r_0^2}{n-m}\left[
\left(\frac{r_0}{r}\right)^{m+1}
-
\left(\frac{r_0}{r}\right)^{n+1}\right] $$

The point where the bond breaks is at $F_{max}$, which can be determined such that

$$\frac{dF}{dr}|_{r=r_{max}} = \frac{E_0r_0^2}{n-m}
\left[
-(m+1)\left(\frac{r_0^{m+1}}{r_{max}^{m+2}}\right)
+
(n+1)\left(\frac{r_0^{n+1}}{r_{max}^{n+2}}\right)
\right]=0 $$

lets related $r_0$ to $r_{max}$ via strain $\delta$ so

$$r_{max} = \delta r_0 $$

and

$$\delta = \left(\frac{n+1}{m+1}\right)^{\frac{1}{n-m}}$$

$\delta$ is typically ~12% to 18%

More importantly, we are now related the single bond strength only to the differnce betwen the attractive and repulsive parameters:



```python
import sys
sys.path.append("/pithy/code")
from pithy import *
%matplotlib inline

n = 12
m = linspace(1,11.9,100)
delta = ((n+1)/(m+1))**(1/(n-m))

plot(m,delta)
ylabel("$\delta$")
xlabel("m")
```

we can relate to stress, by dividing by an area element based on r, so

$$\sigma = \frac{E_0}{n-m}\left[
\left(\frac{r_0}{r}\right)^{m+1}
-
\left(\frac{r_0}{r}\right)^{n+1}\right] $$

and

$$\sigma_{max} = \frac{E_0}{n-m}\left[
\left(\frac{m+1}{n+1}\right)^\frac{m+1}{n-m}
-
\left(\frac{m+1}{n+1}\right)^\frac{n+1}{n-m}
\right] $$


```python
for n in [12,10]:
    fac = (m+1)/(n+1)
    sigmabyE = ((1)/(n-m))*((fac**((m+1)/(n-m)))-(fac**((n+1)/(n-m))))
    plot(m,sigmabyE,label="n=%i"%n)
ylabel("$\\frac{\sigma_{max}}{E_0}$")
xlabel("m")
legend()
print ""
```

$$\rho = \frac{b^2}{c} $$

where the greatest stress is at $x = c$

$$\sigma_c \equiv \sigma_{yy}(c) 
= \sigma_A \left(1 + 2\frac{c}{b} \right)
= \sigma_A \left(1 + 2\sqrt{\frac{c}{\rho}} \right)
$$

which is approximately

$$\sigma_c \approx 2 \sigma_A \sqrt{\frac{c}{\rho}}  $$

This gives us a sense of how "sharp" a crack must as a function of the _real_ failure condition.



```python
c = logspace(-9.2,-5,100)
r = 0.36e-9
sigma_max = 2.5e9
sigma_af = sigma_max*(sqrt(r/c))
plot(c,sigma_af)
loglog()
xlabel("Critical Flaw Size(m)")
ylabel("Applied Stress (Pa)")
```

We see that as the flaw gets "bigger", the stress required to initiate critical failure is much lower, this is best visualized as a ratio: 

```python
c = logspace(-9.2,-5,100)
r = 0.36e-9
sigma_max = 2.5e9
ratio = (sqrt(r/c))
plot(c,ratio)
loglog()
xlabel("Critical Flaw Size(m)")
ylabel("Applied Stress/Theoretical Max")
```

So Flaws that are only ~1 m in size can reduce the effective strength of a material by 2 orders of magnitude!

## Energy Balance During a Crack


We can consider the energy of elastic strain as:

$$ dU_E = \frac{FdL}{AL} = \sigma_x d\epsilon_x $$

$$U_E = \frac{1}{2}E\epsilon_x^2 = \frac{1}{2}\sigma_x \epsilon_x$$

If we have a crack with existing strain $U_E$ and we apply $U_A$, then the total strain on the system $U_M$ is

$$U_M = U_E + U_A$$

Let's relate $U_E$ to $c$, with a crack length of $2c$, so we can calcalate $U_E$:

$$U_E = \frac{\pi c^2 \sigma_A^2}{E}$$

When the crack forms, a new surface with energy $U_s$ will be formed, so

$$U_ = 4c\gamma$$

So the total energy is now

$$U = U_S + U_M $$

where the crack is stable if 

$$\frac{dU}{dc} = 0 $$ 

If the the stress on the plate is stable, then 

$$U_A = 0$$ and
$$U_M = -U_E$$
so
$$\frac{dU_E}{dc} = -\frac{dU_S}{dc}$$

so we can reduce the applied stress required for critial crack growth to

$$\sigma_A = \frac{2\gamma E}{\pi c} \approx 0.8 \sqrt{\frac{\gamma E}{c}} $$


Why is this a more useful parameter?

It is also interesting to note that this relation relates the following

$$\sigma_a \sqrt{c} = \sqrt{\gamma\epsilon} $$

Why is this interesting?

## Composites

To enhance mechanical properties, we often create composite structures. While the mechanical properties of a composite are strongly dependent of the geometry of the composite, we can model a few simple systems to get a sense of what can happen when we combine properties.


For a series stack:

$$\epsilon_x = V\epsilon_{x1} + (1-V)\epsilon_{x2} $$

by Hooke's law:

$$\frac{\sigma_x}{E} = V\frac{\sigma_{x1}}{E_1} 
+ (1-V)\frac{\sigma_{x2}}{E_2} $$


but

$$\sigma_x = \sigma_{x1} = \sigma_{x2} $$

so 

$$\frac{1}{E} = V\frac{1}{E_1} 
+ (1-V)\frac{1}{E_2} $$
