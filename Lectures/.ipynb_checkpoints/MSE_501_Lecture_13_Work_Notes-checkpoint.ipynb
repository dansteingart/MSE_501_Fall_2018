### The Energy At a Surface
---
In the last lecture we motivated that thermodynamics can explain how a phase separated shape can occur, but we sort of fudged the "interface" penalty. Let's explore that in more detail today.

Let's start with the canonical example of soap film.



The work to extend the film can be expressed as

$$W = 2 \gamma dA = 2 \gamma l dx $$

We multiply by 2 in this case because the surface tension $\gamma$ only accounts for one air/soap interface, but there are two such interfaces. 

Q: What is the route of sourface tension?

A: Back to basics


So in many ways the surface can be considered a third phase, since it has an inherenetly different bonding structure from the bulk soap or the bulk film. In our phase field model this is what the order parameter models, if it's a sharp boundary is is just one extra phase, and if it is a diffuse boundary it is an interphase region. Whether or not ths interphase is is considered a 3rd (or 4th or 5th) phase is a matter of definition. I generaly visualize it as a it approaches large thickness as just a mixture of A in B moving gradually to B in A, so it loses definition as something distinct and is therefore no longer a phase. But it is left as an advanced excerise to see if this holds up under a phase rule test.



Since a sharp interface can be considered as a third phase, Young (of the modulus) realized that it can be treated as a membrane of thickness approaching 0, so if we model for  force distribution over the membrane:



we can reduce the thickness to zero and assume at the film center, Scherer `z_s`



A force balance can be decsriubted such that

$$\gamma = \int_0^{z_s} (p''+t) z dz + \int_{z_s}^d (p'+t) z dz $$

where $p'$ and $p''$ are pressures within phase on either side of the arbitrarily defined interface. This is the surface we are describing when we talk about surface tension.

Again, the surface has a real volume, so when we look at the total energy this extensive component needs to be considered and wrapped into our zero-thickness model.




$$n^s_i = n_i - n'_i - n''_i$$


Note that this can be negative! 

We can normalize this to get a surface density, or adsoprtion number 

$$\Gamma_i = \frac{n_i^s}{A}$$

The mol trick we played above applies to all state properties as well, so

$$E^s = E-E'-E''$$

and

$$S^s = S-S'-S''$$

So we can expand to see that this will create additional work _against_ explasion of the system:

$$dW = p'dV'+p''dV''-\gamma dA$$ 

There is only one interface in this model, so $\gamma$ can stand alone. 

And we can add this penalty to an internal energy change:

$$dE = Tds - pdV + \gamma dA$$ 

and enthalpy

$$dE = Tds + Vdp + \gamma dA$$ 

and finally gibbs

$$dG = -SdT + Vdp + \gamma dA$$ 

And assuming constant pressure and temperature:
$$\left(\frac{dG}{dA}\right)_{T,p} = \gamma$$

Now we can add this new energy to our surface!

 


## Gibbs-Duhem Equation, Redux
---

Recall the original surface free relation, and now add surface

$$dG =      \left(\frac{\partial G}{\partial T}\right)_{p,A,n}dT + 
            \left(\frac{\partial G}{\partial p}\right)_{T,A,n}dP + 
\sum^i_{k=1}\left(\frac{\partial G}{\partial n_k}\right)_{T,A,p\neq k}dn +
            \left(\frac{\partial G}{\partial A}\right)_{T,p,n}dA 
$$ 


or 

$$dG = -SdT + VdP + \gamma dA + \sum_{k=1}^i \mu_k dn_k$$

At an interface it's actually practical to assume that dT,dp = 0, so

$$ G = \gamma A + \sum_{k=1}^i \mu_k  n_k$$

And take the full derivative

$$ dG = \gamma dA  +A d\gamma  + \sum_{k=1}^i \mu_k  d n_k + \sum_{k=1}^i d\mu_k  n_k$$

so we can say that

$$ dG - dG = 0 = A d\gamma + \sum_{k=1}^i n_k d\mu_k$$

And since we like to make our interface approach a thickness of 0

$$ dG - dG = 0 = d\gamma + \sum_{k=1}^i \Gamma_k d\mu^s_k$$

so

$$d\gamma = \sum_{k=1}^i \Gamma_k d\mu^s_k $$

let's say that we have one component in our mix that dominates a phase, then we can write

$$\Gamma_1 = -\frac{d\gamma}{d\mu^s_1}$$

Reacalling our functional diulte approximation for $\mu$

$$mu_1^s = RT\ln x_1$$

So

$$\Gamma_1 = -\frac{1}{RT}\frac{d\gamma}{d\ln x_1}$$

Q: What does this tell us?
A: If species one reduces the surface penalty of the system, it will migrate to the interface.

## Back To Solids

### Estimating Surface Energies

Recalling lecture one, we can relate the energy related to a bond to the amouny of energy it takes to escape the potential well, or

$$\frac{2\Delta H_s}{zN_A}$$

Where $\Delta H_s$ is the enthalpy of vaporization, and z is our number of coordinating bonds, and our crystal has $N_A/2$ total bonds.

The surface will be different because of the coordination factor, so

$$\gamma_{cv} = \frac{z-z_s}{z}\frac{N_s\Delta H_s}{N_A}$$

Q: What is the surface cordination in a close packed plane?
A: 4

So in a close packed plane

$$\gamma_{cv} = \frac{N_s\Delta H_s}{4N_A}$$

The same can be done for a liquid, solid interface

Q: Why is the heat of fusion, though, less than the heat of subplimation or varpozation?

A: Because the "new phase" provides fewer compensating bonds. We're alone now.
