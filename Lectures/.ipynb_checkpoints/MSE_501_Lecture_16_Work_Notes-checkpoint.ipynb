## Solid Solid Phase Transformations: Nucleation

$V \Delta G_V + A \gamma_{cc} $ are motivated by out liquid-solid treatment. But liquids support no stress, so they do not "strain". What does this term look like?

### Homogeneous Solid-Solid

From a macroscopic perpective, even when we conserve mass, a volume change can (and often does) accompany a phase change, so we can consider an $\alpha$ to $\beta$ phase transition, with molar volumes $V_{\alpha}$ and  $V_{\beta}$ respectively. 

The volumetric strain caused by the interface between solid phases can be defined as 

$$\epsilon_{\alpha \beta} = \frac{V_{\beta}-V_{\alpha}}{V_{\alpha}} $$


And we can introduce a strain term to our energy balance:

$$ \Delta G = -V \Delta G_V + A \gamma_{cc} + V \Delta G_{s}$$

Assumings isotropic behavior of $\gamma$ with respect to crystal orientation, we can see group

$$\Delta G = -\frac{4}{3}\pi r^3 (\Delta G_V-  \Delta G_{s})+ 4 \pi r^2 \gamma_{cc}$$

With this we see that that, at least initially, the buck strain created by the nucleation of a new phase homegenously within the solid further fights the global thermodynamic drive to form the new phase. Thus, there exists a similar critical $r$, $r^*$, 

$$r^* =  \frac{\gamma}{\Delta G_V- \Delta G_{s}} $$

and thus

$$\Delta G^* =  \frac{16 \pi \gamma^3}{3(\Delta G_V- \Delta G_{s})^2} $$






### Heterogeneous Solid-Solid


Heterogeneous transformation in a solid-solid is akin to the heteregneous solid-liquid case, except we need to define the nature of the heterogenity. 

First, let us classify that anything that isn't _bulk_ $\alpha$ or $\beta$ is a _defect_, and modify $\Delta_G$ such that

$$\Delta G ^{het} = \Delta G - \Delta G_{d} $$ 

Where the negative indicates that the defect provides a "strain relief" to the solid crystal forming. Some examples of the strain relief might include:


- Grain Boundaries
- Vacancies
- Dislocations
- Stacking Faults
- Free Surfaces






What is the order of decreasing "penalty" for solid-solid phase transformation, assuming homogenous is the worst?

Let's think about dimensional order. 

- A vacancy is a 0 dimensional "strain reliever"
- A dislocation is a 1 dimensional "strain reliever"
- A stacking fault is a 1.5 dimensional "strain reliever"
- A grain boundary is a 2 dimensional "strain reliever"
- A free surface is a 3 dimensional "strain reliever"

Rule of thumb: more dimensions equals more places for "excess molar volume" to go, so the easier nucleation becomes as the dimensionality of the defect increases.


## Kinetics of Nucleation

Upon motivating the "extra" driving force required for nucleating, we took a decidely macroscopic approach. But to understand the _rate_ at which nucleation occurs, it's helpful to once again think at the atomic level.

A nuclei forms in the following way:

If we are forming $\beta$ from $\alpha$, we say we have $i$ atoms of $\beta$ joined by one more, so as a reaction

$$\beta_i+ \beta_1 = \beta_{i+1}$$

Taking lot at the microscopic thermodynbamics, we can say that 

$$\Delta U_i = i\Delta \mu + A_i \gamma$$

where $\delta \mu_i$ is the energy difference in the formation of the new phase. $A_i$ is the number of the atoms at the surface and $\gamma$ is the surface energy. If we have an round "embryo", let's not ascribe a macroscopic energy. 

First, derive the number of atoms at the surface (not the surface area!):

$$A_i \approx 4 (i^{1/3})^2$$ 

And the total embryo volume is

$$V_{embryo} = i\Omega $$, 

where  $\omega$ is the molar volume,

$$\Delta G_{vol} = \frac{\Delta \mu}{\Omega} $$

So the surface area of the embryo can be estimated to be

$$A \approx 4 ((i^{1/3})^2) \Omega^{2/3}$$ 


Recall that macroscopic surface energy is 

$$A\gamma_{CL}$$ 

and the microscopic surface energy is 

$$A_i\gamma \approx 4i^{2/3} \gamma $$

So approximating equality between the macroscopic and microscopic terms

$$A\gamma_{CL} \approx   4i^{2/3} \gamma $$ 

So now we can derive a macroscopic variable from microscopic components:

$$ \gamma_{CL} \approx \frac{4i^{2/3}}{A} \gamma =  \frac{\gamma}{\Omega^{2/3}} $$





This manipulations allows us to go back to

$$\Delta U_i = i\Delta \mu + A_i \gamma$$

and say that 

$$\frac{\Delta\mu}{\gamma} \approx \frac{\Omega\Delta G_{vol}}{\Omega^{2/3}\gamma_{CL}} $$


The crystallization surface energy can be hard to measure, but we've related it before to the heat of fusion, and the Turnbull approximation says:


$$N_A \Omega^{2/3} \gamma_{CL} \approx c\Delta H_F = -c V_M T_m \Delta S_{fv}$$

where 

$$c = \frac{1}{3} $$


The heat of fusion is quite easy to measure, as is the melting point, so we can reduce this further and say. Let's say 

$$\Omega^{2/3} \gamma_{CL} \approx -c \Omega T_m \Delta S_{fv} $$


Since: $\Delta G = \Delta S_{fv} \Delta T $,

$$\frac{\Delta\mu}{\gamma} \approx \frac{\Omega\Delta G_{vol}}{\Omega^{2/3}\gamma_{CL}} = 
 \frac{\Omega \Delta S_{fv} \Delta T }{ -c \Omega T_m \Delta S_{fv}} = -\frac{3 \Delta T}{T_m}
$$


So we can now go back to 

$$\Delta U_i = i\Delta \mu + A_i \gamma$$


and say

$$\frac{\Delta U_i}{\gamma} = -i\frac{3\Delta T}{T_m} + 4i^{2/3}$$


```python
import sys
sys.path.append("/pithy/code/")
from pithy import *
%matplotlib inline


def U_by_gamma(i,dT,Tm): return (-i*(3*dT/Tm)+4*i**(2./3))

figure(figsize=(10,10))
n = linspace(0,1e6,20000)
dT = 5
Tm = 273.
plot(n,U_by_gamma(n,dT,Tm),'k')
xlabel('Number of Atoms')
ylabel('$\Delta U / \gamma$')
```

So the critical count is

```python
crit_i = int(n[argmax(U_by_gamma(n,dT,Tm))])
figure(figsize=(10,10))

plot(n,U_by_gamma(n,dT,Tm),'k')
axvline(x=crit_i,color='silver')
xlabel('Number of Atoms')
ylabel('$\Delta U / \gamma$')
print "The critical atom count is %i " % crit_i
```

Going a step further:

```python
dTrange = range(5,100)
crit_size = [int(n[argmax(U_by_gamma(n,dT,Tm))]) for dT in dTrange]

figure(figsize=(10,10))
plot(dTrange,crit_size,'k')
xlabel("Undercooling from Tm = 273")
ylabel("Critical Cluster Count")
semilogy()
print ""
```

So this is how many atoms it takes to form a critical cluster, but how many embryos might exist within a given cluster of atoms?

Well, let's define this as $n_i^e$, where for $n_1$ total atoms.....

$$n_i^e = n_1 \exp [-\Delta U_i / kT] $$

$$\frac{n_i^e}{n_1} =  \exp [-\Delta U_i / kT] \approx \exp[-(-\frac{3\Delta T}{T_m})+4i^{2/3})\frac{c \Delta S/R}{1-\Delta T/T_m}]  $$

assume $c = 1/3$ and $\Delta S/R = 1$, so 

  
 


```python
i = linspace(0,200,100)
c = 1/3.
dT = 1.
Tm = 4
figure(figsize=(10,10))
p1 = -i*(3*dT/Tm)+4*i**(2./3)
p2 = -p1*c/(1-(dT/Tm))
ratio = exp(p2)
plot(i,log(ratio),'k')
ylabel("ln(ne/ntotal)")
xlabel("number of atoms")
print ""
```

So we have a sense the magnitude of the critical atom count as well as the nimber of critical clusters as a fuction of the available atoms. How do we derive a rate from this?


well if we recall

$$\beta_i+ \beta_1 = \beta_{i+1}$$

The we can write a rate equation

$$I_i[t] = n_i[t]r^+_i - n_{i+1}[t]r^-$$

where $r^+$ and $r^-$ are the forward and backwards rate. When the cluster size is constant, I = 0, and at this point


$$n_i[t]r^+_i = n_{i+1}[t]r^-$$



But this is a bit inaccurate, because once we have a critical cluster, it's removed from the "destruction" pool.

So it's helpful to count the rate of change of the number of clusters, such that

$$\frac{dn_i[t]}{dt} = I_{i-1}[t] - I_i[t] $$, where the number clusters for a give size is constant when

$$ I_{i-1}[t] = I_i[t]$$


So the first relation:

$$I_i[t] = n_i[t]r^+_i - n_{i+1}[t]r^-$$

Is only correct when when $i < i^*$, but for all $i$, we can model

$$I_i[t] = n_i^e r^+_i \left[\frac{n_i[t]}{n^e_i}-\frac{n_{i+1}[t]}{n^e_{i+1}} \right] $$


And an analytical approximation for this is dervied by Turnbull as

$$I_v = \frac{kT}{3\pi \lambda^6 \eta}\exp\left({-\frac{\Delta F_i}{kT}}\right)$$

where $\lambda$ is a length scale appoximator and $\eta$ is an effective diffusivity.



So this is generally what determines what the rate, the question now is where does nucleation end and crystal growth begin: next time!
