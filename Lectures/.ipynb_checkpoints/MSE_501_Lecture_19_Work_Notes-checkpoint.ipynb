## Thermoelasticity:

But! The potential well told us something else:

What was it?



This means that Temperature and Elasticity must be related, or at least temperature and stress strain.

From the potential well alone we can write that:

$$\alpha = \frac{1}{L}\frac{dL}{dT} $$

where $\alpha$ is the coefficient of linear thermal eexpansion, and the "free strain" associated with thermal variation is

$$\epsilon_f = \int_{T_0}^T \alpha dT $$

Now, we can go back and add free strain components to our previous stress-strain relationships

$$\epsilon_x = \epsilon_f+\frac{1}{E}\left[\sigma_x-v(\sigma_y+\sigma_z)\right]$$
$$\epsilon_y = \epsilon_f+\frac{1}{E}\left[\sigma_y-v(\sigma_x+\sigma_z)\right]$$
$$\epsilon_z = \epsilon_f+\frac{1}{E}\left[\sigma_z-v(\sigma_x+\sigma_y)\right]$$

By taking this relationship, and mixing with different composite structures, we can model what mismatches in $\alpha$ will do to the stress strate of materials when temperature alone is changed.....
