What's important to note is that _any_ system that crystallize can be forced into a glass state by cooling and or/reacting quickly: is is a question of _how_ fast.

### So How Fast Is Fast?

Roughly speaking, the strutucal complexity of the thermodynamically favored crystal determines how far out the "nose" of the TTT curve is. Metals have relatively simple motifs and repeating units, thus one has to quench standard pure metals and simple alloys excepctionally quickly to have them enter the glassy state. More on this later. 

C-H polymers, as we discussed last time, tend toward amorphous in the same way a length of yarn tends towards knots rather than a ball or a folded structure: the inherent anisotropy in structure and bonding helps. And the converse is true for metals: the simple and isotropic behavior makes it hard to form a glass. So what makes a glass?

So how do we identify a glass?

I really like Scherer here:

>A glass is a solid material lacking long-range order. As indicated schematically in Figure 1, the local environment of the atoms may be the same in the glassy and crystalline forms of a given substance, but the glass does not possess translational symmetry. Glasses exist in every class of liquids, including oxides, metalloids, polymers, and molecular liquids.

And What makes it easy to form a glass?

- High Viscosity at the melting point
- Complex molecular shape
- Complex and varied composition
- Liquid composition far from any simple structure

Cribbing from our discussion of polymerization, if we chemical react the system at a temperature where the viscosity is very high, we can create a confused struture,

Example: CVD

$$ SiCl_4 + O_2 \rightarrow SiO_2 + 2Cl_2 $$

Example: Sol-Gel

$$ Si(OC_2H_5)_4+ 4H_2O \rightarrow Si(OH)_4 + 4C_2H_5_OH $$

then

$$ 2SiOH \rightarrow SiO + H_2O $$

This hydrolysis-condensation chain leads to long polymers. The different is that the SiO bond is much stronger than the C-H bond in air, so when the silica polymer is heated it can "flow" hyperlocally without decomposing, forming a more dense glass

The thermodynamic behavior of the glass is strongly dependent on its history for this reason, and it can be difficult for a single ordering parameter to tell us how it might behave (ala spinodal decomposition) because their can be only one globally stable state, but many metastable states.
