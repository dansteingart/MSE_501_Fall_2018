Let's start with heterogeneous growth from a gas of atoms to a solid: this way we can ignore, for the time being, interaction in the gas phase and sub-surface behavior on the solid.



What drives the atom to the surface of the solid? Well, the same forces that drive bonding, in particular the long range forces of ionic interactions and net-dipoles.

Our atom/molecule in the gas phase has no net charge, but even so small fluctuations in the electron shell provide a net dipole that wiggles on the order of $10^{16} Hz$




We can model this with our lecture one friend:

$$U_a = -\frac{a}{x^6}$$

But we're not just attaching to one atom, we're attracted to the entire solid, so focusing on one atom:




So we can average the interaction by integrating:

$$U(D) = \rho \int_D^\inf \int_0^\inf \left(-\frac{a}{\sqrt{x^2+z+^2}^6} \right) 2\pi x dx dz$$

Where $\rho$ is the density of atoms in the condensed phase.  This becomes

$$ U(D) = - \frac{\pi a \rho}{6D^3} $$

As we can see, the bulk decreases the order of the attractive drop-off from $r^6$ to $r^3$, indicating a longer range interaction than two atoms alone.   


## How Long Does it Take?

Once the atom hits the surface, _it might not stick around forever_. 



To understand this intermediate position between landing and _bonding_, it is helpful to consider a residence time $\tau$, where

$$ \tau = \tau_0 \exp \left(  \frac{Q}{RT} \right) $$

In this scenario $Q$ is our surface adsorption potential well, so


```python
import sys
sys.path.append("/pithy/code/")
from pithy import *
%matplotlib inline

def tau(tau0,Q,T):
    R = 8.3144
    return tau0*exp(Q/(R*T))
figure(figsize=(15,5))
Q = logspace(1,5,500)
T = 298 #K
tau0 = 1e-13

plot(Q,tau(tau0,Q,T),'k')
xlabel("Potential Well (KJ/mol)")
ylabel("Residence Time (arb)")
semilogy()

```

So depending on the nature of the well at the surface, the atom may only last 1 $ps$ (where it bounces off), or it may be effectively "stuck". The order of time approaching $ns$ to $\s s$ is where most interesting things happen for materials growth: longer than $\mu s$ we are effectively stuck and are now a part of the surface. 

## Monolayer Adsorption

Now, let's say we have a group of atoms approaching the surface, and they are stuck to the surface, but they do not interact with one another:



Since the residence time is determined by a single atom-surface interaction, then the number of atoms on the surface at any time should scale with the partial pressure of the atoms in the gas phase (since this is directly related to the number of impingements on the surface per unit time).

We can model this as the ratio of the partial pressure to the temperature of the gas: as pressure increases for a fixed temperature, more atoms should be hitting the surface, but if pressure and temperature increase in lockstep, then we have fewer atoms hitting, just with higher energy.

$$ n = \frac{N_A p}{\sqrt{2\pi M_w RT}} \equiv kp$$

where $N_A$ is Avogadro's number, $M_W$ is the molecular weight of the gas, $T$ is the temperature, $R$ the ideal gas constant. 


Let's take $\theta_0$ as the fraction of open sites on the surface, and $\theta_1$ as the filled fraction. 



The number of atoms the adsorb can be modeled:

$$ n_{ads} = k p \theta_0 A_1 $$

where, if there are not free sites, nothing can stick.

And the rate of desorption can be modeled as  

$$ n_{des} = N_m \theta_1 v_1 \exp \left(-\frac{E_1}{RT}\right) $$


where $N_m$ is the number of atoms required to form a complete monolayer, $v$ is the vibrational frequency of each atom, and $E_1$ represents the energy barrier to detach (akin to $Q$) above.

At equilibrium, there is no net gain or loss of atoms, so:

$$ k p \theta_0 A_1  = N_m \theta_1 v_1 \exp \left(-\frac{E_1}{RT}\right) $$

and surface fraction covered can be expressed as 

$$ \theta_1 = \frac{Kp}{1+Kp} $$

where 

$$ K = \frac{k A_1}{N_m v_1 \exp \left(-\frac{E_1}{RT}\right)} $$

So we have simple but to first order reasonable model that says: for a perfect surface, the adsorbate - surface interaction will be much stronger than the adsorbate-adsorbate interaction, and as such, the surface should completely fill before an adsorbate can land and remain, for any significant time, upon another adsorbate. 

But, this ignores:

- the nature of the absorbate
- the actual structure of the surface




## Multilayer Adsorption

What happens when we have a less than perfect surface?

 


We may have some local "peaks". How to we figure out the tendency for this to happen?  Brunauer-Emmett-Teller (BET) Theory provides a framework for this.




For each layer, we can write:

$$ k p \theta_0 A_1  = N_m \theta_1 v_1 \exp \left(-\frac{E_1}{RT}\right) $$
$$ k p \theta_1 A_2  = N_m \theta_2 v_2 \exp \left(-\frac{E_2}{RT}\right) $$
$$ k p \theta_{n-1} A_n  = N_m \theta_n v_n \exp \left(-\frac{E_n}{RT}\right) $$

This is formally, correct, but we can assume that there only a difference in the 2nd layer and beyond, so


$$ k p \theta_{n-1} A_2  = N_m \theta_2 v_2 \exp \left(-\frac{E_2}{RT}\right) $$

So, let's now calculate the population ratios for each layer

For the first layer,

$$\frac{\theta_1}{\theta_0} = \frac{kpA_1}{N_m v_1 \exp \left(-\frac{E_1}{RT}\right) } \equiv \alpha $$

And for the second:

$$\frac{\theta_2}{\theta_1} = \frac{kpA_2}{N_m v_2 \exp \left(-\frac{E_2}{RT}\right) } \equiv \beta $$

And for all others:

$$\frac{\theta_n}{\theta_{n-1}} = \frac{kpA_2}{N_m v_2 \exp \left(-\frac{E_2}{RT}\right) } = \beta $$

Working backwards:

$$\theta_1 = \alpha \theta_0 $$
$$\theta_2 = \beta \theta_1 = \alpha \beta \theta_0 $$
$$\theta_3 = \beta \theta_2 = \alpha \beta \beta \theta_0 = \alpha \beta^2 \theta_0 $$

so for higher n,

$$\theta_n = \beta \theta_{n-1} = \alpha \beta^{n-1} \theta_0 $$

And we can then calculate the total number of atoms adsorbed, as

$$N = \theta_1 N_m + 2 \theta_2 N_m + 3 \theta_3 N_m +....+ n\theta_n N_m$$

$$N = N_m \sum_{j=1}^n j\theta_j = \alpha\theta_0 N_m \sum_{j=1}^n j\beta^{j-1}$$

So, the fraction of a monolayer adsorbed can be rearranged from this as:


$$\frac{N}{N_m} =\alpha\theta_0 \sum_{j=1}^n j\beta^{j-1}$$


This part is tricky: the area fractions must sum to unity:

$$ 1 = \theta_0 + \sum_{j=1}^n \theta_j = \theta_0 + \alpha \theta_0  \sum_{j=1}^n\beta^{j-1} $$

(why is this the case?)

hint: 


Rearrange to solve for $\theta_0$,

$$\theta_0 = \frac{1}{1+\alpha  \sum_{j=1}^n\beta^{j-1}} $$

Recalling

$$\frac{N}{N_m} =\alpha\theta_0 \sum_{j=1}^n j\beta^{j-1}$$,

we can write:

$$\frac{N}{N_m} =\frac{\alpha  \sum_{j=1}^n j \beta^{j-1}}{1+ \alpha  \sum_{j=1}^n\beta^{j-1}}$$


We can introduce $C$, where

$$C \equiv \frac{\alpha}{\beta}$$,

or a ratio of the relative tendency towards growing or filling. 

So
$$\frac{N}{N_m} =\frac{C  \sum_{j=1}^n j \beta^{j}}{1+ C \sum_{j=1}^n\beta^{j}}$$

We can expand this to realize that

$$\frac{N}{N_m} =\frac{C\beta}{(1-\beta)(1-\beta+C\beta)}$$

We can also now compare the moles to the mass, where

$$\frac{N}{N_m} = \frac{W}{W_m} $$

If we rewind all the way to

$$\frac{\theta_2}{\theta_1} = \frac{kpA_2}{N_m v_2 \exp \left(-\frac{E_2}{RT}\right) } \equiv \beta $$

We can now write

$$1 = \frac{kp_0A_2}{N_m v_2 \exp \left(-\frac{E_2}{RT}\right) } $$

if we define $p_0$ as the condensation condition, so we can consider $\beta$ now as

$$\beta = \frac{p}{p_0} $$

or the ratio of the partial pressure to the condensation equilibrium pressure.

So, we have gone from something we cannot really measure to, this:

$$\frac{W}{W_m} = \frac{C \frac{p}{p_0}}{(1-\frac{p}{p_0})(1-\frac{p}{p_0}+C\frac{p}{p_0})} $$

Where $W$, $W_m$, $p$ and $p_0$ can be known.





There implications here. What this tells is _the amount of a given species that will form a single layer before another layer forms_, not directly _the conditions to form one complete layer_. 

But, because the the former is deterministic, we can use it for all sorts of applications. For example:


$$\theta_0|_{W=W_m}  = \frac{1}{1+\sqrt{C}}$$

This directly tells us the amount of free initial surface which is left available  when the second surface forms. 

In the case of $N_2$, C = 100, so


```python
C_N2 = 100.
theta_0_N2 = 1/(1+sqrt(C_N2))
print str(theta_0_N2 *100)[:4]+"%"
```

So, when a monolayer is formed that will push to a condensed phase of $N_2$, ~10% of the surface is open. Measuring the mass of %N_2% adsorbed as a function of its partial pressure is the standard BET test and is used directly to measure surface area of objects.
