## Composites

To enhance mechanical properties, we often create composite structures. While the mechanical properties of a composite are strongly dependent of the geometry of the composite, we can model a few simple systems to get a sense of what can happen when we combine properties.


For a series stack:

$$\epsilon_x = V\epsilon_{x1} + (1-V)\epsilon_{x2} $$

by Hooke's law:

$$\frac{\sigma_x}{E} = V\frac{\sigma_{x1}}{E_1} 
+ (1-V)\frac{\sigma_{x2}}{E_2} $$


but

$$\sigma_x = \sigma_{x1} = \sigma_{x2} $$

so 

$$\frac{1}{E} = V\frac{1}{E_1} 
+ (1-V)\frac{1}{E_2} $$
